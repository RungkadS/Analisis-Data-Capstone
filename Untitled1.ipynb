```python
weather_recommendations = {
    'Pantai': ['Cerah', 'Cerah Berawan', 'Berawan', 'Berawan Tebal', 'Hujan Ringan'],
    'Gunung': ['Hujan Ringan', 'Cerah', 'Cerah Berawan', 'Hujan Lokal', 'Berawan'],
    'Hutan': ['Hujan Ringan', 'Cerah', 'Cerah Berawan', 'Hujan Sedang', 'Berawan'],
    'Museum': ['Hujan Ringan', 'Cerah', 'Hujan Sedang', 'Cerah Berawan', 'Berawan'],
    'Taman': ['Cerah', 'Cerah Berawan', 'Berawan', 'Kabut', 'Hujan Ringan'],
    'Masjid': ['Hujan Ringan', 'Hujan Sedang', 'Cerah', 'Berawan', 'Hujan Lokal'],
    'Goa': ['Cerah', 'Cerah Berawan', 'Berawan', 'Kabut', 'Hujan Ringan'],
    'Gereja': ['Hujan Ringan', 'Hujan Sedang', 'Cerah', 'Berawan', 'Hujan Lokal'],
    'Candi': ['Cerah', 'Hujan Sedang', 'Hujan Ringan', 'Cerah Berawan', 'Berawan'],
    'Monumen': ['Hujan Ringan', 'Berawan', 'Cerah', 'Hujan Sedang', 'Cerah Berawan'],
    'Kampung': ['Hujan Ringan', 'Hujan Sedang', 'Berawan', 'Cerah', 'Hujan Lokal'],
    'Tugu': ['Hujan Ringan', 'Cerah', 'Berawan', 'Hujan Sedang', 'Cerah Berawan'],
    'Situs': ['Hujan Ringan', 'Cerah', 'Berawan', 'Hujan Sedang', 'Cerah Berawan'],
    'Klenteng': ['Hujan Ringan', 'Hujan Sedang', 'Cerah', 'Berawan', 'Hujan Lokal'],
    'Curug': ['Cerah', 'Berawan', 'Cerah Berawan', 'Kabut', 'Hujan Ringan'],
    'Kebun': ['Cerah', 'Hujan Ringan', 'Berawan', 'Cerah Berawan', 'Hujan Sedang']
}

def get_weather_recommendations(place_name):
    for key, recommendations in weather_recommendations.items():
        if key.lower() in place_name.lower():
            return recommendations
    return ['Cerah', 'Cerah Berawan', 'Berawan', 'Hujan Ringan', 'Hujan Lebat']

merged_df[['Rekomendasi Cuaca 1', 'Rekomendasi Cuaca 2', 'Rekomendasi Cuaca 3',
           'Rekomendasi Cuaca 4', 'Rekomendasi Cuaca 5']] = merged_df['Place_Name'].apply(
    lambda x: pd.Series(get_weather_recommendations(x))
)
```
